## Abstract

In this project, we visit some of the Disk Scheduling algorithms and compare them against each other based on the distance travelled by the head. As parameters, we will use randomly generated disk queues. We will compare six different scheduling algorithms: FCFS, SSTF, SCAN, CSCAN, LOOK, and CLOOK, and we will see the advantages of each one of them,and their respective disadvantages.  

## Results and Discussion

We wish to consider several disk schedulers in our algorithms. First, we need to establish the metric for what a good disk scheduler looks like. We will consider the total distance travelled by the head of the scheduler (seek time) and though there are other things to take into account for disk schedulers, we will stick with seek time. Our queue consist of a large number of requests ranging between 0 and 200. First, we will test out our schedulers to ensure that they work. We will use our disk visualization file and the queue [98,183,37,122,14,124,65,67] starting at 53:

```python
q = [98,183,37,122,14,124,65,67]
head = 53
```

```python
def visualize(func):   
    ds = func(q, head)
    ds.execute()
    disk_visualization.visualize(ds.order)
    print("seek time: ", ds.seek_time)
```

Below, we have FCFS, an algorithm that goes by first come first serve. As expected,this algorithm is not very efficient, since it goes to the next request on the queue regardless of how far it is from our current location. As such, we can expect extremely high seek times. Below, the algorithm gets a seek time of 640, which is pretty high as we'll see. 

```python
visualize(fc)
```

Next up, we have shortest seek time first which simply does the job closest to our current destination first. Admittedly, this algorithm is much better than FCFS, but having to find the closest job to our current position every time is going to add some complexity to the algorithm. Going with our queue above, we are able to reduce the seek time down to 236, almost three times better than FCFS. 

```python
visualize(ss)
```

Scan is another great algorithm that goes from one end to the other, doing all jobs in between. This algorithm is great in that we do not need to look for the next closest process in our queue every time. We get a seek time of 236 using this algorithm. 

```python
visualize(sc)
```

Circular Scan is a sibling of scan, but instead of going back after reaching an end of the disk, it goes back to the beginning and begins to move forward again. This is a pretty algorithm, but it does add some seek time since it goes back every time it reaches the end. Understandably, it performs worse than scan at a seek time of 384. 

```python
visualize(csc)
```

Look is very similar to scan, but instead, look does not go all the way to the end of the disk and instead stops at the highest-valued request and then turns back and proceeds to the lowest-valued request. Starvation can be high for values on the extreme ends especially values opposite of the section in which the head moves towards. However, it is easy to see how look would have a better performance than Scan and thus lower seek times. 

```python
visualize(lk)
```

Clook is the circular version of look, we can expect a higher seek time than look but a lower seek time than cscan. 

```python
visualize(clk)
```

```python
def random_queue(func): 
    q = [random.randint(1,200) for i in range(10000)]
    head = random.randint(1,200)
    ds = func(q, head)
    ds.execute()
    print("seek time: ", ds.seek_time)
```

Below, we see a small simulation that shows just how bad FCFS is - given an infinite-sized queue, scan, look, and their siblings along with sstf, all perform within a bound (below 400, double the size of the disk), but fcfs performs worse as the size of the queue increases with the seek time being unbounded (665563) for a queue of size 10,000. What would happen if we had a finite queue, say a size of 10? If we perform continuously (that is, as we process a demand, we put the next item in the queue), then it turns out that all queues will do rather horribly including look. However, if we process 10 items at first, then the next 10 items, and so on, then we can expect cscan, clook, sstf, look, and scan all to have a seek time of atmost 400, and if our total demands is $x$ big, we can expect a seek time of at most $\frac{400*x}{10} = 40*x$. However, in fcfs, we can expect a seek time of $200*10\frac{x}{10} = 400*x$, so that fcfs does 10 times worse than any other algorithm, at worse. 

```python
random_queue(csc)
```

```python
random_queue(sc)
```

```python
random_queue(fc)
```

```python
random_queue(ss)
```
